### ***Question: 02   Compare regional sales (North America, Europe, Japan, and Rest of the World) for each genre to understand market preferences.***

```python
data = df1.copy()

# Define the regions
regions = ['North America', 'Europe', 'Japan', 'Rest of World']

# Convert regional sales columns to numeric safely
for region in regions:
    data.loc[:, region] = pd.to_numeric(data[region], errors='coerce')

# Group by "Genre" and calculate total sales for each region
regional_sales = data.groupby('Genre')[regions].sum()

# Add a column for total sales across all regions for each genre
regional_sales['Total'] = regional_sales.sum(axis=1)

# Sort the data by total sales for better visualization
regional_sales = regional_sales.sort_values(by='Total', ascending=False)

# Display the result
print("Regional Sales by Genre:")
regional_sales
```

### ***Question: 03  Find the publisher with the highest total global sales.***

```python
 # Group by "Publisher" and calculate total global sales
publisher_sales = df1.groupby('Publisher')['Global'].sum()

# Find the publisher with the highest total global sales
top_publisher = publisher_sales.idxmax()
top_sales = publisher_sales.max()

# Display the result
print(f"The publisher with the highest total global sales is '{top_publisher}' with sales of {top_sales}.")
```

### ***Question: 08 Compare the sales growth trend for "Action" and "Shooter" genres over the years.***

```python
#Add total_region_sales column
df1['Total Regional Sales'] = df1[['North America', 'Europe', 'Japan', 'Rest of World']].sum(axis=1)
df1
```

```python
# Filter for Action and Shooter genres
filtered_df = df1[df1['Genre'].isin(['Action', 'Shooter'])]

# Group by Year and Genre, summing up the global sales
sales_trend = filtered_df.groupby(['Year', 'Genre'])['Total Regional Sales'].sum().reset_index()

# Pivot the data for better visualization
sales_pivot = sales_trend.pivot(index='Year', columns='Genre', values='Total Regional Sales')

# Plot the sales trends
sales_pivot.plot(kind='line', marker='o', figsize=(10, 6))
plt.title("Sales Growth Trend for Action and Shooter Genres")
plt.xlabel("Year")
plt.ylabel("Total Total Regional Sales (in millions)")
plt.legend(title="Genre")
plt.grid()
plt.show()
```

### ***Question: 09 Identify the top publisher in terms of sales for each region.***

```python
regions = ['North America', 'Europe', 'Japan', 'Rest of World']

top_publishers = {}


for region in regions:
    publishers_sales = df1.groupby('Publisher')[region].sum()

    # Identify the top publisher for each region
    top_publisher = publishers_sales.idxmax()
    max_sales = publisher_sales.max()
    top_publishers[region] = {'Publisher' : top_publisher, 'Sales' : max_sales} 
    
# for region, info in top_publishers.items():
#     print(f"Region {region}, Top Publisher: {info['Publisher']} with sales {info['Sales']}")

# Convert results into table display
result_df = pd.DataFrame.from_dict(top_publishers, orient='index')
result_df.index.name = 'Region'
result_df.reset_index(inplace=True)

# Display the table
print(result_df)
```

### ***Question: 11  Create visualizations to better understand and communicate data insights:.***

### **1- Distribution of global sales (histogram).**

```python
plt.figure(figsize=(10, 6))

# Using Seaborn for a histogram with KDE
sns.histplot(df1['Global'], bins=30, kde=False, color='blue')

# Adding titles and labels
plt.title('Distribution of Global Sales', fontsize=16)
plt.xlabel('Global Sales ', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

### The distribution is heavily right-skewed, indicating that a large number of games have low global sales.

### **2- Total sales comparison for regions (bar chart)**

```python
# Calculate total sales for each region
regions = ['North America', 'Europe', 'Japan', 'Rest of World']
total_sales = [df1[region].sum() for region in regions]


plt.figure(figsize=(10, 6))
plt.bar(regions, total_sales, color=['blue', 'orange', 'green', 'purple'])
plt.title('Total Sales Comparison by Region', fontsize=16)
plt.xlabel('Regions', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

for i, value in enumerate(total_sales):
    plt.text(i, value + 0.5, f'{value:.2f}', ha='center', fontsize=10)

# Show the plot
plt.show()
```

### The graph indicates that Europe has the highest sales, followed by North America, while Japan records the lowest sales among the regions. This suggests a stronger market presence in Europe compared to other regions, with Japan contributing relatively little to the overall sales.

### **3- Share of total global sales by genre (pie chart).**

```python
genre_sales = df1.groupby('Genre')['Global'].sum()

# Create the pie chart with custom colors
plt.figure(figsize=(8, 8))
genre_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90,  wedgeprops={'edgecolor': 'white'})
plt.title('Total Global Sales by Genre', fontsize=16)
plt.ylabel('')  # Removes y-axis label for a cleaner look

# Show the plot
plt.show()
```

### The pie chart reveals that the "Action" genre leads in global sales, followed by "Shooter" as the second-highest contributor. Meanwhile, other genres account for significantly smaller portions of the total sales.

### **4- Trend of global sales over the years (line chart).**

```python
yearly_sales = df1.groupby('Year')['Global'].sum()

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('Trend of Global Sales Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Global Sales', fontsize=12)
plt.grid(alpha=0.5)

# Highlight data points
for x, y in zip(yearly_sales.index, yearly_sales.values):
    plt.text(x, y + 0.5, f'{y:.1f}', ha='center', fontsize=8)

# Show the plot
plt.tight_layout()
plt.show()
```

### The graph indicates that sales were highest in the three years 2015, 2016, and 2017, while the lowest sales were recorded in 2018, with sales dropping even further in 2019 and 2020.

### **4-  Top 10 best-selling games globally (horizontal bar chart).**

```python
top_10_games = df1.sort_values(by='Global', ascending=False).head(10)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_games['Game'], top_10_games['Global'], color='skyblue', edgecolor='black')
plt.title('Top 10 Best-Selling Games Globally', fontsize=16)
plt.xlabel('Global Sales', fontsize=12)
plt.ylabel('Games', fontsize=12)

# Add data labels to the bars
for index, value in enumerate(top_10_games['Global']):
    plt.text(value + 0.5, index, f'{value:.2f}', va='center', fontsize=10)

# Invert y-axis to display the highest value at the top
plt.gca().invert_yaxis()

# Show the plot
plt.tight_layout()
plt.show()
```

### **5-  Correlation between regional sales (heatmap).**

```python
regional_sales = df1[['North America', 'Europe', 'Japan', 'Rest of World']]

# correlation
correlation_matrix = regional_sales.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between Regional Sales', fontsize=16)

# Show the heatmap
plt.tight_layout()
plt.show()
```

### The heatmap indicates that Japan has a low correlation with North America, Europe, and the Rest of the World, suggesting distinct sales trends in this region. In contrast, the highest correlation is observed between North America, Europe, and the Rest of the World.

### **6-  Contribution of each genre to regional sales (stacked bar chart).**

```python
genre_sales = df1.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()

# stacked bar chart
genre_sales.plot(kind='bar', stacked=True, figsize=(16, 8), colormap='tab20c')
plt.title('Contribution of Each Genre to Regional Sales', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(title='Regions', loc='upper right', fontsize=10)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

### The graph indicates that the genres 'Action' and 'Shooter' have the highest sales, with Europe contributing the largest share, followed by North America. In contrast, genres such as 'Party,' 'Strategy,' and 'Visual Novel' have the lowest sales across all regions."
